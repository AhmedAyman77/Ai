<h1>Work with videos</h1>


```python
cap = cv2.VideoCapture(0)  # record video
# or
# cap = cv2.VideoCapture("path to your video")

# determining how frames are compressed and stored
fourcc = cv2.VideoWriter_fourcc(*"XVID")

out = cv2.VideoWriter('output.avi', fourcc, 20.0,
                      (640, 480))  # video name, fourcc, frames (how many photos i want to show per second), (width,height)

print(cap.isOpened())  # check if there is something wrong

while cap.isOpened():
    ret, frame = cap.read()  # return ret (bool) , frame if ret = False there is no frame
    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        out.write(frame)  # save frame

        # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # cv2.imshow('frame', gray)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release()  # close camera
out.release()  # close video
cv2.destroyAllWindows()
```
